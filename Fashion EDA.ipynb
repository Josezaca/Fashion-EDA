```python
def extract_videos(json_obj):
    dict_videos = {"channelTitle":[], "channelId":[], "title":[], "publishTime":[], "videoId":[]}
    for i in json_obj["items"]:
        dict_videos["channelTitle"].append(i["snippet"]["channelTitle"])
        dict_videos["channelId"].append(i["snippet"]["channelId"])
        dict_videos["title"].append(i["snippet"]["title"])
        dict_videos["publishTime"].append(i["snippet"]["publishTime"])
        dict_videos["videoId"].append(i["id"]["videoId"])
    df_videos = pd.DataFrame(dict_videos)
    return df_videos
```

```python
dict_videos_2 = {'videoId': [],
                 'viewCount':[],
                 'likeCount':[],
                 'favoriteCount':[],
                 'commentCount':[]}

for i in IDs:
    request_11 = requests.get(f'https://www.googleapis.com/youtube/v3/videos?key={api_key_y}&part=statistics&id={i}')
    json_request = request_11.json()['items'][0]['statistics']
    print(json_request)
    dict_videos_2['videoId'].append(i)
    dict_videos_2['viewCount'].append(json_request.get('viewCount', None))
    dict_videos_2['likeCount'].append(json_request.get('likeCount', None))
    dict_videos_2['favoriteCount'].append(json_request.get('favoriteCount', None))
    dict_videos_2['commentCount'].append(json_request.get('commentCount', None))

df_videos_2 = pd.DataFrame(dict_videos_2)
```

```python

df_copy = df.copy()
df_copy.loc[:, 'likeCount'] = pd.to_numeric(df_copy.loc[:, 'likeCount'])
df_copy.loc[:, 'viewCount'] = pd.to_numeric(df_copy.loc[:, 'viewCount'])
df_copy.loc[:, 'commentCount'] = pd.to_numeric(df_copy.loc[:, 'commentCount'])

df_copy.loc[:, 'likes_log'] = np.log(df_copy.loc[:, 'likeCount'] + 1)
df_copy.loc[:, 'views_log'] = np.log(df_copy.loc[:, 'viewCount'] + 1)
df_copy.loc[:, 'comment_log'] = np.log(df_copy.loc[:, 'commentCount'] + 1)


plt.figure(figsize=(10,5))

plt.subplot(221)
g1 = sns.distplot(df_copy.loc[:, 'views_log'],color='purple')
g1.set_title("VIEWS LOG DISTRIBUTION", fontsize=12)

plt.subplot(223) 
g4 = sns.distplot(df_copy.loc[:, 'comment_log'],color='orange')
g4.set_title("COMMENTS LOG DISTRIBUTION", fontsize=12)

plt.subplot(222) 
g2 = sns.distplot(df_copy.loc[:, 'likes_log'],color='red')
g2.set_title('LIKES LOG DISTRIBUTION', fontsize=12)

plt.subplots_adjust(wspace=0.2, hspace=0.4, top=0.9)

plt.show()
```
